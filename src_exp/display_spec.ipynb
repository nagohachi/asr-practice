```python
from pathlib import Path

root_dir = Path.cwd().parent
input_dir = root_dir / "input"
csj_dir = input_dir / "csj"
spec_npy_dir = csj_dir / "spec_npy"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(csj_dir / "text.train.char.csv")
```

```python
df = df[:10]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

for target, wav_file_path in zip(df["target"], df["source_path"]):
    spec_npy_path = spec_npy_dir / f"{Path(wav_file_path).stem}.npy"
    spec = np.load(spec_npy_path)
    plt.imshow(spec.T, aspect="auto", origin="lower")
    plt.title(target)
    plt.show()
```
